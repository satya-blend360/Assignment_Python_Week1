```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Load the cleaned data
df = pd.read_csv('Cleaned_Amazon_Sale_Report.csv')

# Ensure proper data types
df['Date'] = pd.to_datetime(df['Date'])
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')
 
print("COMPREHENSIVE KPI ANALYSIS - RETAIL SALES DATA") 
print(f"Analysis Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Total Records Analyzed: {len(df):,}") 
```

```python
# KPI 1: Monthly Revenue
# ============================================
monthly_revenue = df.groupby(['Year', 'Month'])['Amount'].sum().reset_index()
monthly_revenue.columns = ['Year', 'Month', 'Total_Revenue']
print("Monthly Revenue:")
print(monthly_revenue)
 
```

```python
# KPI 2: Region-wise (State-wise) Sales 
region_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).reset_index()
region_sales.columns = ['State', 'Total_Sales']
print("Top 10 States by Sales:")
print(region_sales.head(10)) 
```

```python
# KPI 3: Average Order Value (AOV) 
aov = df[df['Amount'] > 0]['Amount'].mean()
print(f"Average Order Value: {aov:.2f}") 
```

```python
# KPI 4: Profit Margin (Assuming 35% margin) 
df['Cost'] = df['Amount'] * 0.65  # 65% of revenue is cost
df['Profit'] = df['Amount'] - df['Cost']
df['Profit_Margin_%'] = (df['Profit'] / df['Amount']) * 100
df['Profit_Margin_%'].fillna(0, inplace=True)

avg_profit_margin = df[df['Amount'] > 0]['Profit_Margin_%'].mean()
print(f"Average Profit Margin: {avg_profit_margin:.2f}%") 
```

```python
# KPI 5: Cancellation Rate 
total_orders = len(df)
cancelled_orders = df['Status'].str.contains('Cancel', case=False, na=False).sum()
cancellation_rate = (cancelled_orders / total_orders) * 100
print(f"Cancellation Rate: {cancellation_rate:.2f}%") 
```

```python
# KPI 6: Top Selling Categories 
category_sales = df.groupby('Category')['Qty'].sum().sort_values(ascending=False).head(10)
print("Top 10 Categories by Quantity Sold:")
print(category_sales) 
 
```

```python
# KPI 7: B2B vs B2C Sales Distribution 
b2b_sales = df.groupby('B2B')['Amount'].sum()
print("B2B vs B2C Sales:")
print(b2b_sales) 
```
