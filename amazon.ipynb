```python
import pandas as pd
import numpy as np

# Display options
 
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', 200)

# Change the file path as needed
df = pd.read_csv("Amazon Sale Report.csv")

df.head()
 
```

```python
df.head()
```

```python
# remove this no need
drop_cols = ["index", "Unnamed: 22"]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])
```

```python
# date was object
df['Date']=pd.to_datetime(df['Date'], errors='coerce')
```

```python
num_cols = ["Qty", "Amount", "ship-postal-code"]

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")  # convert everything to numeric
    df[col] = df[col].fillna(0)

df["ship-postal-code"] = df["ship-postal-code"].astype("Int64")  # fix postal codes
df.info()
```

```python
df["B2B"] = df["B2B"].astype(str).str.lower().map({"true": True, "false": False})
```

```python
# Remove Duplicate Rows

df = df.drop_duplicates()
print("Duplicates removed!")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
print(df.columns.tolist())
```

```python
# df = df.rename(columns={"Sales Channel ": "Sales Channel"})

df.columns = df.columns.str.strip()
```

```python
text_cols = [
    "Status","Fulfilment","Sales Channel","ship-service-level","Style","SKU",
    "Category","Size","ASIN","Courier Status","currency","ship-city",
    "ship-state","ship-country","promotion-ids","fulfilled-by"
]

for col in text_cols:
    df[col] = df[col].astype(str).replace(["nan", "None", ""], np.nan)
    df[col] = df[col].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df["Courier Status"] = df["Courier Status"].replace("Unknown", "Pending")
df["currency"] = df["currency"].replace("Unknown", "INR")
df["promotion-ids"] = df["promotion-ids"].replace("Unknown", "No Promotion")
df['fulfilled-by'] = df['fulfilled-by'].replace("Unknown", np.nan)

df.loc[df['Fulfilment'] == "Merchant", 'fulfilled-by'] = "Merchant"
df.loc[df['Fulfilment'] == "Amazon", 'fulfilled-by'] = "Amazon"
df.loc[df['Fulfilment'].str.contains("Easy", case=False), 'fulfilled-by'] = "Easy Ship"

df["fulfilled-by"] = df["fulfilled-by"].fillna("Unknown")
```

```python
df['ship-city'] = df['ship-city'].str.title()
df['ship-state'] = df['ship-state'].str.title()
df['ship-country'] = df['ship-country'].str.upper()  # IN, US, etc.

df['Category'] = df['Category'].str.title()
df['Size'] = df['Size'].str.upper()
```

```python
df.loc[df["Status"].str.contains("Cancel", case=False), "Amount"] = 0
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["MonthName"] = df["Date"].dt.strftime("%b")
df["Day"] = df["Date"].dt.day
```

```python
df = df.drop_duplicates()
```
